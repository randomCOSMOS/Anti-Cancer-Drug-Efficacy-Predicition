# Importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Iterating through all dataset and plotting their graphs

```python
import os
dir = "./drug data"
values = {}
for file in os.listdir(dir):
    data = pd.read_csv(dir + "/" + file)
    file = file.replace(".csv", "")
    values[file] = [data["IC50"], data["AUC"]]
```

```python
fig, axis = plt.subplots(5,2,figsize=(10,15))
plt.subplots_adjust(top=1.5)
r = 0
c = 0
for i in values:
    axis[r,c].plot(values[i][0],values[i][1])
    axis[r,c].set_title(str(i))
    if r == 4:
        r = 0
        c = c + 1
    else:
        r = r + 1
    
```
